```python
import pandas as pd
```

```python
tables = pd.read_html('https://en.wikipedia.org/wiki/List_of_presidents_of_the_United_States')
tables
```

```python
df = tables[0]
df.head()
```

```python
df.columns
```

```python
df = df.drop(columns=['Portrait'])
```

```python
df = df.drop(columns=['Party[b][17]'])
```

```python
df = df.rename(columns={'Term[16]': 'Term'})
df = df.rename(columns={'Vice President[18]': 'Vice President'})
df = df.rename(columns={'No.[a]': 'No.'})
```

```python
df = df.rename(columns={'Party[b][17].1': 'Party'})
df.head()
```

```python
df.to_csv('List_of_presidents_of_the_United_States.csv', index=False)
```
