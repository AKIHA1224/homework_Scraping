```python
import requests
from bs4 import BeautifulSoup
```

```python
from dotenv import load_dotenv
import os

load_dotenv()

user_agent = os.getenv("USER_AGENT")
```

```python
urls = [
    "https://www.tnwb.uscourts.gov/Search/search.aspx?zoom_query=car&zoom_page=1&zoom_per_page=100&zoom_and=1&zoom_sort=0&zoom_xml=0",
    "https://www.tnwb.uscourts.gov/Search/search.aspx?zoom_query=car&zoom_page=2&zoom_per_page=100&zoom_and=1&zoom_sort=0&zoom_xml=0"
]
rows = []
```

```python
for url in urls:
    response = requests.get(url)
    headers = {"User-Agent":user_agent}
    response = requests.get(url, headers=headers)
    doc = BeautifulSoup(response.text, "html.parser")
    blocks = doc.select('div.result_block, div.result_altblock')
    
    for block in blocks:
        title_tag = block.select_one('div.result_title a')
        case_url = title_tag['href'] if title_tag else ''
        case_name = title_tag.get_text(strip=True) if title_tag else ''
        
        category_tag = block.select_one('span.category')
        category = category_tag.get_text(strip=True).strip("[]") if category_tag else ''

        info_tag = block.select_one('div.infoline')
        supplement = info_tag.get_text(strip=True) if info_tag else ''

        rows.append({
        'Case Name': case_name,
        'Case URL': case_url,
        'Category': category,
        'Additional Details': supplement
})
```

```python
len(rows)
```

```python
import pandas as pd
df = pd.json_normalize(rows)
df.head()
```

```python
df.to_csv("homework7_4CAR.csv", index=False)
```
